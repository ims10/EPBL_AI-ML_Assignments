```python
# prompt: Matplotlib: Draw 10 graphs with proper meaning

import matplotlib.pyplot as plt
import numpy as np

# Sample data for 10 different plots
x = np.linspace(0, 10, 100)

# Create a figure and a set of subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))  # Adjust figsize as needed

# Plot 1: Simple Line Plot
axes[0, 0].plot(x, np.sin(x))
axes[0, 0].set_title("Sine Wave")
axes[0, 0].set_xlabel("x")
axes[0, 0].set_ylabel("sin(x)")


# Plot 2: Scatter Plot
axes[0, 1].scatter(x, np.random.rand(100), color='red', marker='o')
axes[0, 1].set_title("Random Scatter Plot")
axes[0, 1].set_xlabel("x")
axes[0, 1].set_ylabel("Random Values")


# Plot 3: Bar Chart
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 40, 15, 30, 20]
axes[1, 0].bar(categories, values, color='skyblue')
axes[1, 0].set_title("Bar Chart of Categories")
axes[1, 0].set_xlabel("Categories")
axes[1, 0].set_ylabel("Values")

# Plot 4: Histogram
data = np.random.randn(1000)
axes[1, 1].hist(data, bins=30, color='green', alpha=0.7)
axes[1, 1].set_title("Normal Distribution Histogram")
axes[1, 1].set_xlabel("Value")
axes[1, 1].set_ylabel("Frequency")


# Plot 5: Pie Chart
labels = ['Slice 1', 'Slice 2', 'Slice 3']
sizes = [30, 45, 25]
axes[2, 0].pie(sizes, labels=labels, autopct='%1.1f%%', startangle=90)
axes[2, 0].axis('equal')  # Equal aspect ratio ensures that pie is drawn as a circle
axes[2, 0].set_title("Pie Chart Example")


# Plot 6:  Exponential Growth
axes[2, 1].plot(x, np.exp(x/5))
axes[2,1].set_title("Exponential Growth")
axes[2,1].set_xlabel("Time")
axes[2,1].set_ylabel("Growth")


# Plot 7:  Logarithmic Plot
axes[3, 0].plot(x, np.log(x+1)) # Adding 1 to avoid log(0) error
axes[3,0].set_title("Logarithmic Plot")
axes[3,0].set_xlabel("x")
axes[3,0].set_ylabel("log(x)")

#Plot 8:  Errorbars
x_err = np.linspace(0,5,10)
y_err = x_err**2
error = np.random.rand(10)

axes[3,1].errorbar(x_err, y_err, yerr=error, fmt='-o')
axes[3,1].set_title("Error Bars Example")
axes[3,1].set_xlabel("x")
axes[3,1].set_ylabel("y")

#Plot 9:  Cosine wave
axes[4,0].plot(x, np.cos(x), color='purple')
axes[4,0].set_title("Cosine Wave")
axes[4,0].set_xlabel('x')
axes[4,0].set_ylabel('cos(x)')

#Plot 10:  Stem Plot
axes[4,1].stem(x[:10], np.random.randint(1,10,10))
axes[4,1].set_title('Stem Plot')
axes[4,1].set_xlabel('Index')
axes[4,1].set_ylabel('Value')


# Adjust layout to prevent overlapping titles and labels
plt.tight_layout()

# Display the plot
plt.show()
```

# Seaborn: Draw 10 graphs with proper meaning


```python
# prompt: Seaborn: Draw 10 graphs with proper meaning

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Sample DataFrames for different plot types (replace with your actual data)
np.random.seed(0)
n = 100
df = pd.DataFrame({
    'x': np.linspace(0, 10, n),
    'y': np.sin(np.linspace(0, 10, n)) + np.random.normal(0, 0.2, n),
    'category': np.random.choice(['A', 'B', 'C'], n),
    'value': np.random.randint(1, 100, n),
    'group': np.random.choice(['X', 'Y'], n)
})

# Create a figure and axes for 10 subplots
fig, axes = plt.subplots(nrows=5, ncols=2, figsize=(15, 20))

# 1. Scatter Plot
sns.scatterplot(x='x', y='y', hue='category', data=df, ax=axes[0, 0])
axes[0, 0].set_title('Scatter Plot with Hue')

# 2. Line Plot
sns.lineplot(x='x', y='y', data=df, ax=axes[0, 1])
axes[0, 1].set_title('Line Plot')

# 3. Histogram
sns.histplot(df['value'], kde=True, ax=axes[1, 0])
axes[1, 0].set_title('Histogram with KDE')

# 4. Bar Plot
sns.countplot(x='category', data=df, ax=axes[1, 1])
axes[1, 1].set_title('Count Plot')

# 5. Box Plot
sns.boxplot(x='category', y='value', data=df, ax=axes[2, 0])
axes[2, 0].set_title('Box Plot')

# 6. Violin Plot
sns.violinplot(x='category', y='value', data=df, ax=axes[2, 1])
axes[2, 1].set_title('Violin Plot')

# 7. Heatmap (using a correlation matrix)
corr = df[['x', 'y', 'value']].corr()
sns.heatmap(corr, annot=True, cmap='coolwarm', ax=axes[3, 0])
axes[3, 0].set_title('Correlation Heatmap')

# 8. Pair Plot
sns.pairplot(df, hue='category', vars=['x', 'y', 'value']) # Removed ax=axes[3,1]
#Removed axes[3,1].set_title('Pairplot (check separate figure)') #Title won't appear on this subplot due to pairplot nature.
plt.show() # Show the pairplot in a new window.


# 9. Joint Plot
sns.jointplot(x='x', y='y', data=df, kind='kde')
plt.suptitle('Joint Plot (check separate figure)', y=1.02) # title for JointPlot


# 10. Swarm Plot
sns.swarmplot(x='group', y='value', data=df, ax=axes[4, 0])
axes[4, 0].set_title('Swarm Plot')

# Remove empty subplot
axes[4,1].axis('off')

# Improve layout
plt.tight_layout()
plt.show()
```
